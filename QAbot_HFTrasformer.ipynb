```python
import torch 
from transformers import pipeline
import pandas as pd
```

```python
tqa = pipeline(task='table-question-answering', model='google/tapas-base-finetuned-wtq')
```

```python
df = pd.read_parquet("dataClean.parquet")
```

```python
df = df.astype(str)
```

```python
query = 'on which day I had the highest value of Active Energy Burned?'
print(tqa(table=df.head(20), query=query)['answer'])
```
